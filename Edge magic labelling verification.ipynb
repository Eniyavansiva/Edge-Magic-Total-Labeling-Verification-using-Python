
#### Edge-Magic Total Labeling Verification

Author: S. Eniyavan
M.Sc. Mathematics Thesis Work

This script implements and verifies constructive edge-magic total
labelings for:

1. Triangular Book Graph B(3,n)
2. Bistar Graph S_{m,n}

Verification includes:
- Bijection onto {1,...,|V|+|E|}
- Constant edge sum validation

This implementation provides computational support
for the theoretical constructions derived in the thesis.


```python
# ============================================================
# Edge-Magic Total Labeling Verification Script
# ============================================================

# ------------------------------------------------------------
# Triangular Book Graph B(3,n)
# ------------------------------------------------------------

def verify_triangular_book(n):
    """
    Verifies edge-magic total labeling for B(3,n).
    Returns verification results.
    """
    
    v = n + 2
    e = 2*n + 1
    total_labels = v + e   # = 3(n+1)
    
    labels = {}
    
    # ----- Vertex Labels -----
    labels["v1"] = 1
    labels["v2"] = n + 2
    
    for i in range(1, n+1):
        labels[f"v{i+2}"] = i + 1
    
    # ----- Edge Labels -----
    labels["v1v2"] = 2*n + 3
    
    for i in range(1, n+1):
        labels[f"v1v{i+2}"] = 3*n + 4 - i
        labels[f"v{i+2}v2"] = 2*n + 3 - i
    
    # ----- Magic Constant -----
    k = 3*n + 6
    
    # ----- Check Bijection -----
    used = set(labels.values())
    is_bijection = (used == set(range(1, total_labels + 1)))
    
    # ----- Check Edge Sums -----
    correct_sum = True
    
    # Spine edge
    if labels["v1"] + labels["v2"] + labels["v1v2"] != k:
        correct_sum = False
    
    # Page edges
    for i in range(1, n+1):
        if labels["v1"] + labels[f"v{i+2}"] + labels[f"v1v{i+2}"] != k:
            correct_sum = False
        
        if labels[f"v{i+2}"] + labels["v2"] + labels[f"v{i+2}v2"] != k:
            correct_sum = False
    
    return {
        "n": n,
        "magic_constant": k,
        "bijection": is_bijection,
        "constant_edge_sum": correct_sum
    }


# ------------------------------------------------------------
# Bistar Graph S_{m,n}
# ------------------------------------------------------------

def verify_bistar(m, n):
    """
    Verifies edge-magic total labeling for S_{m,n}
    using symmetric algebraic construction.
    """
    
    v = m + n + 2
    e = m + n + 1
    total_labels = v + e   # = 2(m+n)+3
    
    labels = {}
    
    # ----- Vertex Labels -----
    labels["u"] = 1
    labels["v"] = m + n + 2
    
    for j in range(1, m+1):
        labels[f"u{j}"] = 1 + j
    
    for i in range(1, n+1):
        labels[f"v{i}"] = 1 + m + i
    
    # ----- Magic Constant -----
    k = 3*m + 2*n + 5
    
    # ----- Edge Labels (forced algebraically) -----
    labels["uv"] = k - labels["u"] - labels["v"]
    
    for j in range(1, m+1):
        labels[f"u-u{j}"] = k - labels["u"] - labels[f"u{j}"]
    
    for i in range(1, n+1):
        labels[f"v-v{i}"] = k - labels["v"] - labels[f"v{i}"]
    
    # ----- Check Bijection -----
    used = set(labels.values())
    is_bijection = (used == set(range(1, total_labels + 1)))
    
    # ----- Check Edge Sums -----
    correct_sum = True
    
    if labels["u"] + labels["v"] + labels["uv"] != k:
        correct_sum = False
    
    for j in range(1, m+1):
        if labels["u"] + labels[f"u{j}"] + labels[f"u-u{j}"] != k:
            correct_sum = False
    
    for i in range(1, n+1):
        if labels["v"] + labels[f"v{i}"] + labels[f"v-v{i}"] != k:
            correct_sum = False
    
    return {
        "m": m,
        "n": n,
        "magic_constant": k,
        "bijection": is_bijection,
        "constant_edge_sum": correct_sum
    }


# ------------------------------------------------------------
# Example Testing
# ------------------------------------------------------------

print("Triangular Book Tests")
for n in range(2, 8):
    print(verify_triangular_book(n))

print("\nBistar Tests")
for m in range(2, 6):
    for n in range(2, 6):
        print(verify_bistar(m, n))
```
